```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
def plot_graph(linespace, funcs_list, labels_list, title=None):
    n = np.linspace(1, linespace, linespace)
    for func, label in zip(funcs_list, labels_list):
        plt.plot(n, func(n), label=label)
    plt.xlabel("Number of Inputs")
    plt.ylabel("Time")
    plt.legend(bbox_to_anchor=(1, 1))
    plt.title(title)
    plt.show()
```

# 1.1

```python
def f1(n):
    return 10000 * n

def f2(n):
    return 10000 * n * np.log2(n)

def f3(n):
    return 100 * n ** 2

def f4(n):
    return 10 * n ** 3

def f5(n):
    return 2 ** n

funcs_list = [f1, f2, f3]
labels_list = ["10000n", "10000n log n", "100n^2"]
title = "All Funcs w/o f4(n) and f5(n)"
plot_graph(1500, funcs_list, labels_list, title)

funcs_list = [f1, f2, f3, f4]
labels_list = ["10000n", "10000n log n", "100n^2", "10n^3"]
title = "All Funcs w/o f5(n)"
plot_graph(1500, funcs_list, labels_list, title)

funcs_list = [f1, f2, f3, f4, f5]
labels_list = ["10000n", "10000n log n", "100n^2", "10n^3", "2^n"]
title = "All Funcs"
plot_graph(30, funcs_list, labels_list, title)
```

```python
def find_2nd_largest_element(arr):
    first, second = max(arr[0], arr[1]), min(arr[0], arr[1])
    for i in range(2, len(arr)):
        if second < arr[i]:
            if first < arr[i]:
                second = first
                first = arr[i]
            else:
                second = arr[i]
    return second
```

**time: O(n)  space: O(1)**

```python
num_of_elems = [3, 5, 10]
test_cases = []
for i, _ in enumerate(num_of_elems):
    test_cases.append([random.randint(0, 20) for _ in range(num_of_elems[i])])

for case in test_cases:
    print("case:", case)
    print("2nd elem:", find_2nd_largest_element(case))
```
